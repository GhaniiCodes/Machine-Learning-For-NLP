# N - Grams :-->

An n-gram is a sequence of n adjacent symbols, which could be letters, syllables, words, or even phonemes or base pairs

```python
# Importing Required Libraries

import pandas as pd
import re
import nltk
from nltk.stem import WordNetLemmatizer
```

Initializing the Count vectorizer and Lemmatizer

```python
Lemmatizer = WordNetLemmatizer()
Messages = []
Stop_Words = ["is" ,"that", "this", "the" , "a" , "i" , "if" , "it", "did" , "can" , "with", "but" , "for" , "to" , "in",
              "his" , "our" , "and" , "will" , "on"]
```

Reading the Data

```python
Data = pd.read_csv("Spam Classification.csv")
```

```python
Data.head()
```

```python
Data.tail()
```

```python
Data.describe()
```

```python
for i in range (len(Data)):
    Cleaned = re.sub("[^a-zA-Z]" , " ", Data['Message'][i])
    Cleaned = Cleaned.lower()
    Cleaned = Cleaned.split()
    Lemmatized = [Lemmatizer.lemmatize(word) for word in Cleaned if not word in Stop_Words]
    Cleaned = " ".join(Lemmatized)
    Messages.append(Cleaned)
```

```python
Messages
```

```python
from sklearn.feature_extraction.text import CountVectorizer
Count_Vectorizer = CountVectorizer(max_features=200, binary=True)

BOW = Count_Vectorizer.fit_transform(Messages).toarray()
BOW
```

It will show the indexes of the words

```python
Count_Vectorizer.vocabulary_
```

```python
from sklearn.feature_extraction.text import CountVectorizer
Count_Vectorizer = CountVectorizer(max_features=1000, binary=True ,ngram_range=(1,2))

BOW = Count_Vectorizer.fit_transform(Messages).toarray()
BOW
```

```python
Count_Vectorizer.vocabulary_
```

```python
from sklearn.feature_extraction.text import CountVectorizer
Count_Vectorizer = CountVectorizer(max_features=1000, binary=True ,ngram_range=(1,3))

BOW = Count_Vectorizer.fit_transform(Messages).toarray()
BOW
```

```python
Count_Vectorizer.vocabulary_
```

### Here's a more detailed breakdown of the advantages:


##### Contextual Understanding:
N-grams capture contextual meaning by considering the relationships between words in a sequence. This allows for a more nuanced understanding of text compared to analyzing individual words in isolation. 

##### Predictive Power:
N-gram models can predict the likelihood of a word appearing based on the preceding words, making them useful in tasks like text prediction and language generation. 

##### Ease of Implementation:
N-grams are relatively simple to implement and understand, as they involve counting the frequency of word sequences. 

##### Versatile Applications:
N-grams find use in various fields, including speech recognition, machine translation, sentiment analysis, and predictive text input. 

##### Improved Accuracy:
By considering context and word relationships, n-grams can improve the accuracy of various NLP tasks, such as part-of-speech tagging, sentiment analysis, and text classification. 

##### Reduced Resource Requirements:
N-gram indexes can reduce the computational resources needed to analyze large text datasets, making it efficient for handling large volumes of data.
